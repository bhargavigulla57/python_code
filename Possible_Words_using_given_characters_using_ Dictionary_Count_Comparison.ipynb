```python
d = ['go', 'bat', 'me', 'eat', 'goal', 'boy', 'run']
ch = ['e', 'o', 'b', 'a', 'm', 'g', 'l']

for word in d:
    valid = True
    for c in word:
        if c not in ch or word.count(c) > ch.count(c):
            valid = False
            break
    if valid:
        print(word)
```
