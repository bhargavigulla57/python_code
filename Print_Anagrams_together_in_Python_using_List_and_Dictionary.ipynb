```python
from collections import defaultdict

a = ['bat', 'nat', 'tan', 'ate', 'eat', 'tea']
res = defaultdict(list)

for word in a:
    key = ''.join(sorted(word))
    res[key].append(word)

print(list(res.values()))
```
