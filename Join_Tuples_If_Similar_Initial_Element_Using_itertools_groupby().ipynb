```python
from itertools import groupby

t = [(5, 6), (5, 7), (6, 8), (6, 10), (7, 13)]
res = []

for k, g in groupby(t, key=lambda x: x[0]):
    vals = [v for _, v in g]
    res.append((k, *vals))

print(res)
```
