```python
from collections import defaultdict

d = {'gfg': [1, 2, 3], 'is': [1, 4], 'best': [4, 2]}
res = defaultdict(list)
for key, vals in d.items():
    for v in vals:
        res[v].append(key)
print(dict(res))
```
